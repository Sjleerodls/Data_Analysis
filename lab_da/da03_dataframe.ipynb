```python
# 데이터프레임에 파생변수(새로운 컬럼) 추가
tips['tip_ration'] = tips.tip / tips.total_bill
```

```python
tips.head()
```

* 팁/영수증 비율의 평균값
* 팁/영수증 비율이 평균보다 큰 데이터
* 팁/영수증 비율이 평균보다 큰 데이터 중에서 남성/여성의 빈도 수

```python
tips.tip_ration.mean()  # 0.16 => 16%
```

```python
tips[tips.tip_ration > tips.tip_ration.mean()]  # 평균보다 큰 107개의 행
```

```python
bigger_ratio = tips[tips.tip_ration > tips.tip_ration.mean()]
bigger_ratio.sex.value_counts()
```

```python
tips.sex.value_counts()
```

```python
bigger_ratio.sex.value_counts() / tips.sex.value_counts()
```
