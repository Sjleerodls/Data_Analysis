## 객실 등급별 생존 비율

```python
titanic.head()
survive = titanic[titanic.alive == 'yes']   # 생존자 DF
survive.pclass.value_counts()
```

```python
titanic.pclass.value_counts()
```

```python
survive.pclass.value_counts() / titanic.pclass.value_counts()
```

## 각 객실 등급(First, Second, Third)에서 남성/ 여성/ 어린이 생존 비율

```python
# 1등급 객실 생존자
first = titanic[(titanic.pclass == 1) & (titanic.survived == 1)]
print(first.who.value_counts())

first_all = titanic[titanic.pclass == 1]
print(first_all.who.value_counts())

res_first = first.who.value_counts() / first_all.who.value_counts()
res_first
```

```python
# 2등급 객실 생존자
second = titanic[(titanic.pclass == 2) & (titanic.survived == 1)]
print(second.who.value_counts())

second_all = titanic[titanic.pclass == 2]
print(second_all.who.value_counts())

res_second = second.who.value_counts() / second_all.who.value_counts()
res_second
```

```python
# 3등급 객실 생존자
third = titanic[(titanic.pclass == 3) & (titanic.survived == 1)]
print(third.who.value_counts())

third_all = titanic[titanic.pclass == 3]
print(third_all.who.value_counts())

res_third = third.who.value_counts() / third_all.who.value_counts()
res_third
```

```python
# for문 사용
for c in sorted(titanic.pclass.unique()):   # sorted()를 사용해 새로운 리스트를 반환. (원본 훼손 x)
    result = titanic[(titanic.pclass == c) & (titanic.survived == 1)].who.value_counts() / titanic[titanic.pclass == c].who.value_counts()
    print(f'--- class {c} ---')
    print(result)
```

```python
# 오쌤
for c in range(1, 4):
    print(f'--- class {c} ---')
    subset = titanic[titanic.pclass == c]   # 객실 등급의 승객
    alive_subset = subset[subset.alive == 'yes']    # 객실에서 생존자
    print(alive_subset.who.value_counts() / subset.who.value_counts())
```
