# Categoryical Data

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(data={
    'user_id': np.arange(1,11),
    'gender':['M']*5 + ['F']*5,
    'score': np.random.randint(1, 6, size=10)
})
```

```python
df
```

```python
df.info()
```

```python
df.gender.value_counts()
```

```python
df.score.value_counts()
```

```python
df.gender = df.gender.astype('category')    # astype : 타입을 명시적으로 변환해주는 함수
```

```python
df.info()
```

```python
df.gender.cat.categories    # 데이터프레임 내 카테고리 타입은 .cat.categories를 통해
                            # 2개의 범주를 갖는 문자열 타입의 인덱스라는 것을 확인할 수 있음.

#> DataFrame.column.cat.categories
```

```python
df.score = df.score.astype('category')
```

```python
df.info()
```

```python
df.score
```

```python
df.score.cat.categories     # 카테고리 접근자 사용 가능.
```

```python
df['score'] = df.score.cat.add_categories(5)
```

```python
df.score.cat.categories
```

```python
df.score
```

# 연속형 변수에서 파생된 카테고리 변수 만들기

```python
df = pd.DataFrame(data={
    'id': np.arange(1, 11),
    'age': np.random.randint(0, 100, size=10)
})
```

```python
df
```

```python
df['age_range'] = pd.cut(x=df.age, bins=np.arange(0,101, 10), right=False)   # cut? 카테고리를 만들어줌
                        # 기본값 right=True 에서 False로 바꿈으로서 x <= age_range < 80으로 변경함.
```

```python
df
```

```python
df.info()
```

```python
df.age_range.cat.categories
```

```python
df['age_cat'] = pd.cut(x=df.age, bins=[0, 19, 40, 60, 100], right=False,
                       labels=['미성년', '청년', '중년', '노년'])
```

```python
df
```

```python
df.age_cat.cat.categories
```
